#### Title : Text Preprocessing - Implement text preprocessing techniques, including tokenization and 
#### normalization, and apply it to any text data.  

```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
text="""Text preprocessing techniques are essential 
for improving the performance of Natural Language Processing models."""
```

```python
text=text.lower()
```

```python
text=text.translate(str.maketrans('','',string.punctuation))
```

```python
tokens=word_tokenize(text)
```

```python
sw=set(stopwords.words("english"))
ft=[word for word in tokens if word not in sw]
```

```python
lem=WordNetLemmatizer()
lt=[lem.lemmatize(word) for word in ft]
```

```python
stemmer=PorterStemmer()
st=[stemmer.stem(word) for word in lt]
```

```python
print("Original Text:\n",text)
print("\n Tokens:\n",tokens)
print("\n After stopword Removal:\n",ft)
print("\n After Lemmatization :\n",lt)
print("\n After Stemming :\n",st)
```

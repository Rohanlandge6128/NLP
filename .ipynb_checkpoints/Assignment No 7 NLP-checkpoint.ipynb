```python
import json

with open(r"C:\Users\Rohan Landge\Downloads\Cell_Phones_and_Accessories_5.json\Cell_Phones_and_Accessories_5.json", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

print("Dataset:", len(data))
```

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))

sentences = [d.get("reviewText","") for d in data]

tokenized_sentences = [
    [w for w in word_tokenize(re.sub(r'[^a-zA-Z ]','', s.lower())) if w not in stop_words]
    for s in sentences if s
]
```

```python
## Word2Vec

from gensim.models import Word2Vec
model=Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=5
)

output=model.wv.most_similar("movie")
for word,score in output:
    print(word,score)
```

```python
model.wv.similarity("movie","movies")
```

```python
## Fast Texting 

from gensim.models import FastText
fasttext_model=FastText(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=5
)

output=fasttext_model.wv.most_similar("movie")
for word,score in output:
    print(word,score)
```

```python
fasttext_model.wv.similarity("movie","movies")
```

#### Lesk Algorithm :

```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
word="bank"
```

```python
sent1="I went to the bank to deposit money."
sent2="The river overflowed the bank after heavy rain."
```

```python
tokens1=word_tokenize(sent1)
tokens2=word_tokenize(sent2)
```

```python
sense1=lesk(tokens1,word)
sense2=lesk(tokens2,word)
```

```python
print("Sentence 1:",sent1)
print("Predicted Sense :",sense1)
print("Meaning :",sense1.definition())
```

```python
print("Sentence 2:",sent2)
print("Predicted Sense:",sense2)
print("Meaning :",sense2.definition())
```

#### Overlapping Score:

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
```

```python
word="bank"
```

```python
sent1="I went to the bank to deposit money."
sent2="The river overflowed the bank after heavy rain."
```

```python
sentences=[sent1,sent2]
```

```python
for sentence in sentences:
    print("\n Sentence :",sentence)
    context_words=set(word_tokenize(sentence.lower()))

    max_overlap=0
    best_sense=None

    for syn in wn.synsets(word):
        signature=syn.definition()

        for ex in syn.examples():
            signature+=" "+ex

        signature_words=set(word_tokenize(signature.lower()))

        overlap=len(context_words.intersection(signature_words))

        print("\n Sense :",syn.name())
        print("Definition :",syn.definition())
        print("Overlap Score:",overlap)

        if overlap>max_overlap:
            max_overlap=overlap
            best_sense=syn

    print("\n Selected Meaning :", best_sense.definition())
```

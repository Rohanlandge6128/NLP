```python
import warnings
warnings.filterwarnings("ignore")

from fastcoref import FCoref

# Load model
model = FCoref(device='cpu')

text = """
Rohit went to the market. He bought apples.
Later, Rohit met Suresh and he invited him for dinner.
"""

# Run prediction
preds = model.predict(texts=[text])

result = preds[0]

# Print clusters
clusters = result.get_clusters(as_strings=True)

print("Coreference Clusters:")
for cluster in clusters:
    print(cluster)

print("\nResolved Text:")

resolved_text = text

# Replace pronouns with first mention of cluster
for cluster in clusters:
    main_entity = cluster[0]   # first mention is main entity
    for mention in cluster[1:]:
        resolved_text = resolved_text.replace(mention, main_entity)

print(resolved_text)
```

```python
import re

text = """Ravi went to the market. He bought apples.
The boy was happy because he found fresh fruit."""

#Extract proper noun
def extract_main_entity(text):
    words = text.split()
    for w in words:
        clean = w.strip(".,")
        if clean[0].isupper() and clean.lower() not in ["the", "a", "an"]:
            return clean
    return "Person"

main_entity = extract_main_entity(text)

#Define reference patterns 
references = [
    "the boy","the girl","the man","the woman",
    "boy","girl","man","woman",
    "he","she","him","her","his"
]

resolved_text = text

#Replace references carefully
for ref in references:
    pattern = r'\b' + ref + r'\b'
    resolved_text = re.sub(pattern, main_entity, resolved_text, flags=re.IGNORECASE)

print("----- Original Text -----\n")
print(text)

print("\n----- Resolved Text -----\n")
print(resolved_text)
```

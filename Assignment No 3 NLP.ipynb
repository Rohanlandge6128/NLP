```python
import numpy as np

states = ['N', 'V']
observations = ['time', 'flies']

# Initial probabilities
pi = {'N': 0.6, 'V': 0.4}

# Transition probabilities
A = {
    'N': {'N': 0.3, 'V': 0.7},
    'V': {'N': 0.8, 'V': 0.2}
}

# Emission probabilities
B = {
    'N': {'time': 0.5, 'flies': 0.2},
    'V': {'time': 0.1, 'flies': 0.6}
}

# Viterbi Algorithm
V = [{}]
path = {}

# Initialization
for state in states:
    V[0][state] = pi[state] * B[state][observations[0]]
    path[state] = [state]

# Recursion
for t in range(1, len(observations)):
    V.append({})
    new_path = {}

    for curr_state in states:
        (prob, prev_state) = max(
            (V[t-1][prev_state] * A[prev_state][curr_state] * B[curr_state][observations[t]], prev_state)
            for prev_state in states
        )

        V[t][curr_state] = prob
        new_path[curr_state] = path[prev_state] + [curr_state]

    path = new_path

# Termination
(n, state) = max((V[len(observations)-1][state], state) for state in states)
best_path = path[state]

print("Most probable tag sequence:", best_path)
print("Probability:", n)
```
